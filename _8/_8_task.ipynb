```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV, train_test_split
```

```python
# гружу датасет
df = pd.read_csv('creditcard.csv')
```

```python
df.head()
```

```python
# подтверждаю несбалансированность выборки
df['Class'].value_counts(normalize=True)
```

```python
# вроде, все норм, все числа, кроме целевой
df.info()
```

```python
pd.options.display.max_columns = 100
```

```python
df.head(10)
```

```python
# создаю X без класса
X = df.drop('Class', axis =1)
X.head(10)
```

```python
# создаю y 
y = df['Class']
type(y)
```

```python
# делаю сплит, чекаю формы
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=100, stratify=y)
print(f'X_train: {X_train.shape}, X_test: {X_test.shape}, y_train: {y_train.shape}, y_test: {y_test.shape}')
```

```python
# параметры сетки 
parameters = [{'n_estimators': [10, 15], 
               'max_features': np.arange(3, 5),
               'max_depth': np.arange(4, 7)}]
```

```python
# завожу модел поиска по сетке
clf = GridSearchCV(estimator=RandomForestClassifier(random_state=100),
                  param_grid=parameters,
                  scoring='roc_auc',
                  cv=3)
```

```python
# обучаю модель
clf.fit(X_train, y_train)
```

```python
# чекаю лучшие параметры
clf.best_params_
```

```python
# вангую вероятности классов
pd.DataFrame(clf.predict_proba(X_test)).head(10)
```

```python
y_pred_proba = clf.predict_proba(X_test)[:, 1]
y_pred_proba
```

```python
from sklearn.metrics import roc_auc_score
```

```python
# считаю ошибку
roc_auc_score(y_test, y_pred_proba)
```
