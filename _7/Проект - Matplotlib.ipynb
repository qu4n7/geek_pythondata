```python
min_price = data['Price'].min()
max_price = data['Price'].max()
```

```python
min_price, max_price
```

```python
ls = np.linspace(min_price, max_price, 21)
```

```python
ls
```

```python
qty_flat = []

for i in range(20):
    if i < 19:
        qty_flat.append(data.loc[(data['Price'] >= ls[i]) & (data['Price'] < ls[i+1]), :].shape[0])
    else:
        qty_flat.append(data.loc[(data['Price'] >= ls[i]) & (data['Price'] <= ls[i+1]), :].shape[0])
```

```python
qty_flat = np.array(qty_flat, dtype=np.float64)
qty_flat
```
