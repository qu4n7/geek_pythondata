```python
x = np.linspace(-5, 5, 101)
x2 = x**2
x3 = x**3
```

```python
plt.axis([-5, 5, -8, 8])
plt.plot(x, x2)
plt.plot(x, x3)
```

```python
fig, ax = plt.subplots(nrows=1, ncols=1)
ax.plot(x, x2)
```

```python
fig, ax = plt.subplots(nrows=1, ncols=2)
ax1, ax2 = ax
ax1.plot(x, x2)
ax2.plot(x, x3)
plt.show()
```

```python
fig, ax = plt.subplots(nrows=1, ncols=2)
ax1, ax2 = ax.flatten()
ax1.plot(x, x2)
ax2.plot(x, x3)
plt.show()
```

```python
fig, ax = plt.subplots(nrows=1, ncols=2)
ax1, ax2 = ax.flatten()

ax1.plot(x, x2)
ax1.set_title('График x**2', color = 'red')
ax1.set_xlabel('x', color ='red')
#ax1.xaxis.label.set_color('red') # alternative to change the color
ax1.tick_params(axis='x', colors='red')
ax1.tick_params(axis='y', colors='red')
ax1.spines['bottom'].set_color('red')
ax1.spines['top'].set_color('red')

ax2.set_title('График x**3', color = 'yellow')
ax2.set_xlabel('x', color ='yellow')
ax2.tick_params(axis='x', colors='yellow')
ax2.tick_params(axis='y', colors='yellow')
ax2.spines['bottom'].set_color('yellow')
ax2.spines['top'].set_color('yellow')
plt.show()
```

```python
fig, ax = plt.subplots(nrows=1, ncols=2)
ax1, ax2 = ax.flatten()
ax1.plot(x, x2)
ax2.plot(x, x3)
fig.set_size_inches(8, 3.5)
```

```python
fig, ax = plt.subplots(nrows=1, ncols=2)
ax1, ax2 = ax.flatten()
ax1.plot(x, x2)
ax2.plot(x, x3)
fig.set_size_inches(8, 3.5)
plt.subplots_adjust(wspace=0.3)
```

```python
fig, ax = plt.subplots(nrows=2, ncols=1)
print(ax.shape)
ax1, ax2 = ax.flatten()
ax1.plot(x, x2)
ax2.plot(x, x3)
fig.set_size_inches(3, 6)
```

```python
y = np.sin(x)
z = np.cos(x)
```

```python
fig, ax = plt.subplots(nrows=2, ncols=2)
ax1, ax2,ax3, ax4 = ax.flatten()
ax1.plot(x, x2)
ax2.plot(x, x3)
ax3.plot(x, y)
ax4.plot(x, z)
fig.set_size_inches(8, 6)
plt.subplots_adjust(wspace=0.3, hspace=0.3)
```

```python
fig, ax = plt.subplots(nrows=2, ncols=2)
ax1, ax2,ax3, ax4 = ax.flatten()

ax1.plot(x, x2)
ax1.set_title('x*2')
ax2.plot(x, x3)
ax2.set_title('x**3')
ax3.plot(x, y)
ax3.set_title('y')
ax4.plot(x, z)
ax4.set_title('z')

fig.set_size_inches(8, 6)
plt.subplots_adjust(wspace=0.3, hspace=0.3)
```

```python
fig, ax = plt.subplots(nrows=2, ncols=2, sharey=True)
ax1, ax2, ax3, ax4 = ax.flatten()

ax1.plot(x, x2)
ax1.set_title('График x**2')

ax2.plot(x, x3)
ax2.set_title('График x**3')

ax3.plot(x, y)
ax3.set_title('График sin(x)')

ax4.plot(x, z)
ax4.set_title('График sin(x)')

fig.set_size_inches(8, 6)
plt.subplots_adjust(wspace=0.3, hspace=0.3)
```

```python
fig, ax = plt.subplots(nrows=2, ncols=2)
ax1, ax2, ax3, ax4 = ax.flatten()

ax1.plot(x, x2)
ax1.set_title('График x**2')
ax1.set_xlim([-3,3])

ax2.plot(x, x3)
ax2.set_title('График x**3')

ax3.plot(x, y)
ax3.set_title('График sin(x)')

ax4.plot(x, z)
ax4.set_title('График sin(x)')

fig.set_size_inches(8, 6)
plt.subplots_adjust(wspace=0.3, hspace=0.3)
```
